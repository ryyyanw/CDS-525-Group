```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Dropout, Bidirectional, Layer
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import backend as K
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
plt.style.use('seaborn-v0_8-white')

# ========================== Hyperparameters ==========================
MAX_VOCAB = 20000
MAX_LEN = 300
EMBEDDING_DIM = 100
LSTM_UNITS = 128
DROPOUT = 0.3
EPOCHS = 6
RANDOM_STATE = 42
GLOVE_PATH = "/kaggle/input/datasets/habibatariq/glove6b100/glove.6B.100d.txt"

# ========================== Dataset load ==========================
df = pd.read_csv("/kaggle/input/datasets/iamrahulthorat/fakenews-csv/fakenews.csv")
df = df.dropna(subset=["text", "label"])
df["content"] = df["text"]

X, y = df["content"].values, df["label"].values
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

# 文本编码
tokenizer = Tokenizer(num_words=MAX_VOCAB, lower=True)
tokenizer.fit_on_texts(X_train)
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)
X_train_pad = pad_sequences(X_train_seq, maxlen=MAX_LEN)
X_test_pad = pad_sequences(X_test_seq, maxlen=MAX_LEN)
word_index = tokenizer.word_index

# 加载 GloVe
def load_glove():
    embeddings_index = {}
    with open(GLOVE_PATH, encoding='utf-8') as f:
        for line in f:
            word, coefs = line.split(maxsplit=1)
            embeddings_index[word] = np.fromstring(coefs, "f", sep=" ")
    embedding_matrix = np.zeros((MAX_VOCAB, EMBEDDING_DIM))
    for word, i in word_index.items():
        if i < MAX_VOCAB and embeddings_index.get(word) is not None:
            embedding_matrix[i] = embeddings_index[word]
    return embedding_matrix

embedding_matrix = load_glove()

# ========================== Attention layer ==========================
class AttentionLayer(Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)
    def build(self, input_shape):
        self.W = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                 initializer="glorot_uniform", trainable=True)
        super().build(input_shape)
    def call(self, x):
        e = K.tanh(K.dot(x, self.W))
        alpha = K.softmax(e, axis=1)
        return K.sum(alpha * x, axis=1)

# ========================== model fit ==========================
def build_model(loss_fn="binary_crossentropy", lr=1e-3):
    inputs = Input(shape=(MAX_LEN,))
    
    embedding = Embedding(
        MAX_VOCAB, EMBEDDING_DIM, weights=[embedding_matrix],
        trainable=True, mask_zero=False
    )(inputs)
    
    bilstm = Bidirectional(LSTM(LSTM_UNITS, return_sequences=True, use_cudnn=False))(embedding)
    bilstm = Dropout(DROPOUT)(bilstm)
    att = AttentionLayer()(bilstm)
    dense = Dense(64, activation="relu")(att)
    dense = Dropout(DROPOUT)(dense)
    
    # los
    if loss_fn == "bce_with_logits":
        output = Dense(1, activation=None)
    else:
        output = Dense(1, activation="sigmoid")
    output = output(dense)
    
    model = Model(inputs, output)
    
    # another loss function switch
    if loss_fn == "bce_with_logits":
        loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)
    else:
        loss = loss_fn
        
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss=loss,
        metrics=["accuracy"]
    )
    return model

early_stop = EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)

# ========================== train models ==========================
def train_model(loss_fn, batch_size, lr):
    model = build_model(loss_fn=loss_fn, lr=lr)
    history = model.fit(
        X_train_pad, y_train,
        batch_size=batch_size, epochs=EPOCHS,
        validation_data=(X_test_pad, y_test),
        callbacks=[early_stop], verbose=1
    )
    return model, history

# ========================== train models ==========================
# base：Binary Crossentropy
model_base, hist_base = train_model(
    loss_fn="binary_crossentropy", batch_size=64, lr=1e-3
)

# compare：BCEWithLogitsLoss
model_bce, hist_bce = train_model(
    loss_fn="bce_with_logits", batch_size=64, lr=1e-3
)

# ========================== plot figures ==========================
plt.figure(figsize=(14, 8))  

# max epochs for each model
epochs_base = range(1, len(hist_base.history['loss']) + 1)
epochs_bce = range(1, len(hist_bce.history['loss']) + 1)

# ---------------- plot BCE model(binary_crossentropy) loss function blue ----------------
plt.plot(epochs_base, hist_base.history['loss'], 'b-o', linewidth=2, markersize=6, label='BCE: Training Loss')
plt.plot(epochs_base, hist_base.history['accuracy'], 'b-s', linewidth=2, markersize=6, label='BCE: Training Accuracy')
plt.plot(epochs_base, hist_base.history['val_accuracy'], 'b--^', linewidth=2, markersize=6, label='BCE: Test Accuracy')

# ---------------- plot BCE model(bce_with_logits) loss function red ----------------
plt.plot(epochs_bce, hist_bce.history['loss'], 'r-o', linewidth=2, markersize=6, label='BCE_with_Logits: Training Loss')
plt.plot(epochs_bce, hist_bce.history['accuracy'], 'r-s', linewidth=2, markersize=6, label='BCE_with_Logits: Training Accuracy')
plt.plot(epochs_bce, hist_bce.history['val_accuracy'], 'r--^', linewidth=2, markersize=6, label='BCE_with_Logits: Test Accuracy')

plt.title('Comparison of Different Loss Functions (Training & Test Metrics)', fontsize=16)
plt.xlabel('Number of Epochs', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.legend(fontsize=11, bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout() # 自动调整布局
plt.savefig("loss_function_comparison.png", dpi=300, bbox_inches='tight')
plt.show()


# ==============================================================================
# different learning rate
# ==============================================================================
lrs = [0.1, 0.01, 0.001, 0.0001]
plt.figure(figsize=(14,10))
for i, lr in enumerate(lrs):
    train_and_plot("binary_crossentropy",64,lr,f"LR={lr}",221+i)
plt.suptitle("Different Learning Rates")
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()

# ==============================================================================
# different batch size
# ==============================================================================
batches = [8,16,32,64,128]
plt.figure(figsize=(14,10))
for i, bs in enumerate(batches):
    train_and_plot("binary_crossentropy", bs, 1e-3, f"Batch Size={bs}",231+i)
plt.suptitle("Different Batch Sizes")
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()

# ==============================================================================
# TOP 100 from test Dataset
# ==============================================================================
def predict_with_logits(model, X):
    logits = model.predict(X, verbose=0)
    return (tf.sigmoid(logits).numpy() > 0.5).astype(int).flatten()

# 基础模型预测
y_pred = (model_base.predict(X_test_pad, verbose=0) > 0.5).astype(int).flatten()

res = pd.DataFrame({
    "Text Preview": [t[:80]+"..." for t in X_test[:100]],
    "True Label": y_test[:100],
    "Pred Label": y_pred[:100],
    "Correct": y_test[:100] == y_pred[:100]
})

print("="*70)
print("Test Set Top 100 Results")
print("="*70)
print(res.to_string(index=False))

plt.figure(figsize=(14,20))
plt.axis('off')
tbl = plt.table(cellText=res.values, colLabels=res.columns, cellLoc='center', loc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(9)
tbl.scale(1.2,2)
plt.title("Top 100 Test Predictions", pad=20)
plt.show()
```
